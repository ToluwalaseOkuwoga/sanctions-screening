### Column Descriptions

- **id:**  
  The unique identifier of the given entity.

- **schema:**  
  Entity type (e.g., Person, Company).

- **name:**  
  The preferred name of the given entity.

- **aliases:**  
  All names (e.g., other scripts, nom de guerre) provided by the data sources.

- **birth_date:**  
  For people, their birth date.

- **countries:**  
  List of countries; includes countries of residence, nationalities, and corporate jurisdictions.

- **addresses:**  
  List of known addresses for the entity.

- **identifiers:**  
  Identifiers such as corporate registrations, passport numbers, or tax identifiers linked to this sanctions target.

- **sanctions:**  
  Details regarding the sanctions designation, if any.

- **phones:**  
  List of phone numbers linked to the entity, in E.164 format.

- **emails:**  
  List of email addresses linked to the entity.

- **dataset:**  
  List of dataset names that provide source data for this entity.

- **first_seen:**  
  Earliest date this entity has been noticed by OpenSanctions.

- **last_seen:**  
  Most recent time this entity was observed in source data.

- **last_change:**  
  Most recent time the checksum of the values in this entity has changed.

```python
# Define a cleaning function for the birth_date strings
def clean_date_str(date_str):
    """
    Clean a date string by stripping whitespace and removing any trailing timezone offset.
    """
    if pd.isna(date_str):
        return date_str
    # Ensure the input is a string and strip whitespace
    date_str = str(date_str).strip()
    # Remove any trailing timezone offset (e.g., "+25:00" or "-25:00")
    cleaned = re.sub(r'([+-]\d{2}:?\d{2})$', '', date_str)
    return cleaned.strip()
```

```python
# Apply the cleaning function to the 'birth_date' column ---
df['birth_date_clean'] = df['birth_date'].apply(clean_date_str)

# Replace literal "nan" strings with proper missing values (np.nan)
df['birth_date_clean'] = df['birth_date_clean'].replace("nan", np.nan)

print(df['birth_date_clean'].head(10).apply(lambda x: repr(x)))
```

```python
# Convert the cleaned column to datetime using an explicit format
# Since the valid dates seem to be in ISO format "YYYY-MM-DD", we force that format.
df['birth_date_parsed'] = pd.to_datetime(
    df['birth_date_clean'].str.strip(), 
    format='%Y-%m-%d', 
    errors='coerce'
)
print("Parsed birth_date values:")
print(df[['birth_date_clean', 'birth_date_parsed']].head(10))
print("\nElement types in 'birth_date_parsed':")
print(df['birth_date_parsed'].head(10).apply(lambda x: type(x)))
print("\nIs the column fully datetime?", pd.api.types.is_datetime64_any_dtype(df['birth_date_parsed']))
```

```python
today = pd.to_datetime(datetime.today())
df['age'] = (today - df['birth_date_parsed']).dt.days // 365

print("\nParsed birth_date and calculated age:")
print(df[['birth_date_parsed', 'age']].head(10))
```

```python
# Function to map an alpha-2 country code to its full country name.
def map_country_code(code):
    try:
        # Ensure the code is uppercase
        country = pycountry.countries.get(alpha_2=code.upper())
        if country:
            return country.name
    except Exception as e:
        pass
    # If not found, return the original code
    return code

# Processing the countries column:
# 1. Fill missing values with an empty string.
# 2. Split the string on ';' to get individual country codes.
# 3. Strip spaces and convert codes to uppercase.
df['country_list'] = df['countries'].fillna('').apply(
    lambda x: [code.strip() for code in x.split(';') if code.strip()]
)

# Map each country code to its full country name.
df['country_names'] = df['country_list'].apply(
    lambda codes: [map_country_code(code) for code in codes]
)

print(df[['countries', 'country_list', 'country_names']].head(10))
```

```python
# Function to normalize a text string: lowercase and remove diacritics.
def normalize_text(s):
    s = s.lower()  # lowercasing
    # Remove diacritics
    s = unicodedata.normalize('NFKD', s)
    s = ''.join(c for c in s if not unicodedata.combining(c))
    return s.strip()

# Process the aliases string:
def process_aliases(aliases_str):
    if pd.isna(aliases_str) or not aliases_str:
        return []
    # Split the aliases by semicolon and strip extra whitespace
    aliases = [a.strip() for a in aliases_str.split(';') if a.strip()]
    # Normalize each alias
    normalized_aliases = [normalize_text(a) for a in aliases]
    # Remove duplicates while preserving order
    seen = set()
    unique_aliases = []
    for alias in normalized_aliases:
        if alias not in seen:
            seen.add(alias)
            unique_aliases.append(alias)
    return unique_aliases

```

```python
df['normalized_aliases'] = df['aliases'].apply(process_aliases)

# Flag records with an excessive number of aliases
# define excessive as having more than 10 aliases
excessive_threshold = 10
df['excessive_alias_flag'] = df['normalized_alias_count'] > excessive_threshold

print("\nRecords flagged for excessive aliases:")
df[df['excessive_alias_flag']][['name', 'aliases','normalized_aliases', 'alias_count', 'normalized_alias_count']].head()
```
